```python
%load_ext autoreload
%autoreload 2

import cv2
import numpy as np
import sys

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [16, 12]

from calibration import *

RESOLUTION = (1920,1440,3)

def captureImage(fname):
    img = cv2.imread(fname)
    if img.shape[:2] != RESOLUTION[:2]:
        img = cv2.resize(img,RESOLUTION[:2])
    return img

def drawImage(img):
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
intrinsic_calibratior = IntrinsicCalibrator()

basepath = os.path.abspath('')
param_path = os.path.join(basepath,'test.yaml')
image_path = os.path.join(basepath,'checker_imgs')

intrinsic_calibratior.paramsFromImageDir(param_path,image_path)
```

```python
img = captureImage(os.path.join(image_path,'checker_4.jpg'))

out = intrinsic_calibratior.applyIntrinsicTransform(img)
drawImage(img)
drawImage(out)
```
